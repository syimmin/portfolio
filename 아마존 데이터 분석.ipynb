```python
import re
import nltk
from nltk.corpus import stopwords
from gensim.parsing.preprocessing import preprocess_string

nltk.download('stopwords')
```

```python
def clean_text(d):
    pattern = r'[^a-zA-Z\s]'
    d = re.sub(pattern, '', d)
    return d
    
def remove_link(d):
    d = re.sub('https?://\S+', '', d)
    return d

def clean_stopword(d):
    stop_words = stopwords.words('english')
    return " ".join([w.lower() for w in d.split() if w not in stop_words and len(w) > 3])

def preprocessing(d):
    return preprocess_string(d)
```

```python
df['review_content'] = df['review_content'].apply(remove_link)
```

```python
df['review_content'] = df['review_content'].apply(clean_text)
```

```python
df['review_content'][0]
```

```python
df['review_content'] = df['review_content'].apply(clean_stopword)
```

```python
df['review_content'][0]
```

```python
tokenized_review = df['review_content'].apply(preprocessing)
tokenized_review = tokenized_review.to_list()
tokenized_review
```
